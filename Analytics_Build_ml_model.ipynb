```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/Healthcare_Data_Preprocessed_FIXED.csv')
data.head()
```

Removing null values

```python
#To remove null values
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = data.select_dtypes(include=['object', 'bool']).columns

for col in numerical_cols:
    data[col] = data[col].fillna(data[col].mean())

for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
data.isnull().sum()
```

```python
#to convert categorical cols into numerical
data['Diet_Type_Vegan']=data['Diet_Type_Vegan'].astype(int)
data['Diet_Type_Vegetarian']=data['Diet_Type_Vegetarian'].astype(int)
data['Blood_Group_AB']=data['Blood_Group_AB'].astype(int)
data['Blood_Group_B']=data['Blood_Group_B'].astype(int)
data['Blood_Group_O']=data['Blood_Group_O'].astype(int)
data
```

```python
correlation_with_target = data.corr()['Target'].sort_values(ascending=False)
print("Correlation with Target variable:\n", correlation_with_target)
```

```python
X=data.drop(columns='Target')
y=data['Target']
print(X.shape)
print(y.shape)
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)

model = sm.Logit(y, X).fit()

model.summary()
```

Removing p>0.05(Not significant features)

```python
features_to_remove = model.pvalues[model.pvalues > 0.05].index
print(features_to_remove)

X = X.drop(columns=features_to_remove, axis = 1)
```

```python
X.shape,y.shape
```

Removing vif>=4

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(data):
    vif_data = pd.DataFrame()
    vif_data["Variable"] = data.columns
    vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    return vif_data

calculate_vif(X)
```

```python
if 'const' in X.columns:
    X = X.drop('const', axis = 1)
vif_threshold = 4 # Applying the requested threshold
max_vif = float('inf')
VIF_features_removed = []

# Iterate and remove features until all VIFs are below the threshold
while max_vif > vif_threshold:
    # Calculate VIF for current X
    vif_df = calculate_vif(X)

    # Find the maximum VIF score among remaining features
    max_vif = vif_df["VIF"].max()

    # If the maximum VIF is above the threshold, remove that variable
    if max_vif > vif_threshold:
        remove = vif_df.sort_values("VIF", ascending=False).iloc[0]

        # Drop the identified variable from X
        VIF_features_removed.append(remove["Variable"])
        X.drop(remove["Variable"], axis=1, inplace=True)

# Display the features and their VIFs after the iterative removal
final_vif = calculate_vif(X)
print("Features removed due to high VIF:", VIF_features_removed)
print("\nFinal VIF scores for remaining features:")
print(final_vif.sort_values(by='VIF', ascending=False))
```

Splitting Data and Model building

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_train_pred = log.predict(X_train)
y_test_pred = log.predict(X_test)
```

```python
from sklearn.metrics import f1_score, classification_report

score_train = f1_score(y_train, y_train_pred)
print('Train F1 score:', score_train)
print('Train Classification Report:\n', classification_report(y_train, y_train_pred))

score_test = f1_score(y_test, y_test_pred)
print('\nTest F1 score:', score_test)
print('Test Classification Report:\n', classification_report(y_test, y_test_pred))
```
